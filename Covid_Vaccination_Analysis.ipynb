### Data Visualization of complete data

### How many entries for each state?

```python
covid19India_df['State/UnionTerritory'].value_counts().sort_values() 
```

```python
covid19India_df_Haryana = covid19India_df[covid19India_df['State/UnionTerritory']=='Haryana']
covid19India_df_Haryana
covid19India_df_Kerala = covid19India_df[covid19India_df['State/UnionTerritory']=='Kerala']
covid19India_df_Kerala
covid19India_df_WestBengal = covid19India_df[covid19India_df['State/UnionTerritory']=='West Bengal']
covid19India_df_WestBengal
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Haryana dataset")

# Line chart  
sns.lineplot(data=covid19India_df_Haryana)
```

```python
covid19India_df_by_State = covid19India_df.groupby('State/UnionTerritory')[['Cured','Deaths' ,'Confirmed']].sum().reset_index()
covid19India_df_by_State
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Cured")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=covid19India_df_by_State.index, y=covid19India_df_by_State['Cured'])

# Add label for vertical axis
plt.ylabel("Cured")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Confirmed")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=covid19India_df_by_State.index, y=covid19India_df_by_State['Confirmed'])

# Add label for vertical axis
plt.ylabel("Confirmed")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Deaths")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=covid19India_df_by_State.index, y=covid19India_df_by_State['Deaths'])

# Add label for vertical axis
plt.ylabel("Deaths")
```
